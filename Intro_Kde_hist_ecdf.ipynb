#More Examples

KDE plots
1. univariate plots

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
# univariate (1D)
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower, shade = True);
```

```python
sns.kdeplot(cars.horsepower, bw_adjust = 0.2);
```

```python
sns.kdeplot(cars.horsepower, cumulative = True);
```

#Bivariate data

```python
sns.set()
```

```python
sns.kdeplot(cars.horsepower, cars.mpg);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20);
```

```python
sns.set_style('whitegrid')
sns.kdeplot(cars.horsepower, cars.mpg, shade = True, shade_lowest= False );
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, shade = True, shade_lowest= False , cbar = True);
```

#HistPlot

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace= True)
```

```python
sns.set()
```

```python
sns.histplot(penguins.flipper_length_mm)
```

```python
sns.histplot(x="flipper_length_mm", data= penguins)
```

```python
sns.histplot(y="flipper_length_mm", data= penguins)
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, kde= True, bins = 20)
```

```python
sns.histplot(x="flipper_length_mm", data= penguins,  bins = 20, stat='density') # aarea of bins is 1 because it is density
```

```python
sns.histplot(x="flipper_length_mm", data= penguins,  bins = 20, stat='probability') # height of bars should sum upto 1
```

```python
sns.histplot(x="flipper_length_mm", data= penguins,  bins = 20, stat='probability', fill=False, element='step', cumulative=True)
```

#Categorical Variables

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species") # hue contains naame of categorical column
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species", element="step")
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species", element="poly")
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species", multiple="layer")
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species", multiple="stack")
```

```python
sns.histplot(x="flipper_length_mm", data= penguins, hue="species", multiple="fill")
```

#Bivariate Histplot

```python
sns.set_style("white")
```

```python
sns.histplot(x="flipper_length_mm", y="bill_length_mm", data=penguins , cbar=True , hue="species")
```

```python
sns.histplot(x="flipper_length_mm", y="species", data=penguins , hue="species")
```

#Styling Tips

```python
sns.histplot(x="species", hue="sex", multiple="dodge", data=penguins, shrink = 0.8 , color="indigo") # dodge property and shrink
```

```python
sns.histplot(x="species", hue="sex", multiple="dodge", data=penguins, shrink = 0.8 , palette='bone')
```

```python
sns.histplot(x="species", hue="sex", multiple="dodge", data=penguins, shrink = 0.8 , color="indigo", fill= False)
```

#Ecdfplot
Empirical cumulative distribution function

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.set_style("darkgrid")
```

```python
# x-axis
sns.ecdfplot(x="tip", data=tips)
```

```python
sns.ecdfplot(x="tip", data=tips )
plt.axvline(4,c='black')
```

```python
sns.ecdfplot(y="tip", data=tips )
plt.axvline(0.5,c='black')
```

```python
tips.tip.median()
```

```python
sns.ecdfplot(x="tip", data=tips , hue="time")
```

```python
sns.ecdfplot(x="tip", data=tips , hue="day")
```

```python
sns.ecdfplot(x="tip", data=tips  , stat="count")
```

```python
sns.ecdfplot(x="tip", data=tips  , stat="count", complementary=True)
```

```python
sns.ecdfplot(x="tip", data=tips  , weights="tip")
plt.axhline(0.5, c="black")
```

```python
sns.ecdfplot(x="tip", data=tips , hue="day", palette="summer")
```

```python
sns.ecdfplot(x="tip", data=tips , hue="day", palette="summer", lw=3)
```

```python
p= sns.ecdfplot(x="tip", data=tips , hue="day", palette="summer", lw=3)
p.legend(["t", "f", "s", "su"],title="DOW", loc=4 )
```
