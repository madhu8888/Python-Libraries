#Boxplot
1. Median : 50 % left , 50% right
2. 25 % on left and right
3. IQR (25-75%)
5. 1.5 * IQR (both sides)
6. Outlier 
7. Summary statistics to understand your data

```python
sns.set()
```

```python
cars = sns.load_dataset("mpg").dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(cars.mpg)
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin , y=cars.mpg)
```

```python
sns.boxplot(x="origin", y="mpg", data=cars)
```

```python
sns.boxplot(x="origin", y="mpg",hue="cylinders", data=cars) # one more category hue
```

```python
# create own category on fly
cars["newer_model"]= cars.model_year >76
```

```python
sns.boxplot(x="origin", y="mpg",hue="newer_model", data=cars)
```

#ViolinPlots
1. 1 box plot
2. 2 kde plot
3. categorical distribution plots
4. median, IQR, whiskers (same)
5. violin plots show kde estimation od pdf

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars)# hue as another category
```

Properties:
1. Symmetric 


```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars[cars.origin.isin(['japan', 'europe'])])
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars[cars.origin.isin(['japan', 'europe'])], split= True) # for splitt property we can only have two categories
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars[cars.origin.isin(['japan', 'europe'])], split= True , inner = "quartile") # for splitt property we can only have two categories
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars[cars.origin.isin(['japan', 'europe'])], split= True , scale = "count") # for splitt property we can only have two categories
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",data=cars[cars.origin.isin(['japan', 'europe'])], split= True , scale = "count", scale_hue= False) # for splitt property we can only have two categories
plt.legend(loc=2)
```

Box options

```python
sns.violinplot(x="origin", y="displacement", data=cars , order=["japan", "europe", "usa"])
```

```python
sns.violinplot(x="origin", y="displacement", data=cars , order=["japan", "europe", "usa"], linewidth=2.5)
```

KDE options

```python
sns.violinplot(x="origin", y="displacement", data=cars , order=["japan", "europe", "usa"], bw= 0.2) # bandwidth
```

#SWARM PLOTS

1. sorting data
2. group by value
3. circles for each point
4. no circle overlaps

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

```python
#filter by cars haaving even number of cylinder
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.swarmplot(x=cars.origin, y = cars.horsepower)
```

```python
sns.swarmplot(x="origin", y = "horsepower" , data=cars)
```

```python
sns.swarmplot(x="origin", y = "horsepower" , hue="cylinders", data=cars)
```

```python
sns.swarmplot(x="origin", y = "horsepower" , hue="cylinders", data=cars, dodge=True)
```

Swarmplot overlay

```python
usa = cars[cars.origin=='usa']
```

with boxplot

```python
sns.boxplot(x=usa.cylinders, y = usa.horsepower)
sns.swarmplot(x=usa.cylinders, y = usa.horsepower)
```

```python
import numpy as np
sns.boxplot(x=usa.cylinders, y = usa.horsepower, whis=np.inf)
sns.swarmplot(x=usa.cylinders, y = usa.horsepower , color="black", alpha=0.5)
```

overlay with violinplot

```python
sns.violinplot(x=usa.cylinders, y = usa.horsepower)
sns.swarmplot(x=usa.cylinders, y = usa.horsepower)
```

```python
sns.violinplot(x=usa.cylinders, y = usa.horsepower, scale="width")
sns.swarmplot(x=usa.cylinders, y = usa.horsepower, color="white")
```

```python
sns.violinplot(x=usa.cylinders, y = usa.horsepower, scale="width")
sns.swarmplot(x=usa.cylinders, y = usa.horsepower, color="white", size=4)
```

```python
sns.violinplot(x=usa.cylinders, y = usa.horsepower, scale="width", inner=None)
sns.swarmplot(x=usa.cylinders, y = usa.horsepower, color="white", size=4)
```

```python
sns.violinplot(x=usa.cylinders, y = usa.horsepower, scale="width", inner=None , color="pink")
sns.swarmplot(x=usa.cylinders, y = usa.horsepower, color="white", size=4)
```

styling

```python
sns.swarmplot(cars.horsepower, linewidth=1)# add line outside each dot
```

```python
sns.swarmplot(cars.horsepower, linewidth=1, edgecolor="black") # black edge color
```

matplotlib scatter properties: marker

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(8,4))
sns.swarmplot(cars.horsepower, marker="*")
```

#STRIPPLOT
1. Jitter
2. Categorical scatter plot, same as swarmplot
3. dots arranged randomly, color intensity to estimate data distribution

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.stripplot(cars.weight);
```

```python
sns.stripplot(cars.weight, cars.origin);
```

```python
sns.stripplot(x="weight", y="origin", data = cars);
```

```python
sns.stripplot(x="weight", y="origin",hue='cylinders', data = cars);
```

```python
sns.stripplot(x="weight", y="origin",hue='cylinders', data = cars, dodge=True);
```

jitter

```python
sns.stripplot(x="weight", y="origin", data = cars, jitter=False);
```

```python
sns.stripplot(x="weight", y="origin", data = cars, jitter=0.05);
```

jitter widget

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
  sns.stripplot(x="weight", y="origin", data = cars, jitter=jitter)
```

```python
interactive(jitter_widget, jitter=(0, 0.65,0.05))
```

alpha

```python
sns.stripplot(x="weight", y="origin", data = cars, alpha=0.5, size=8); # points size and transparency
```

linewidth and edgecolor

```python
sns.stripplot(x="weight", y="origin", data = cars, linewidth=2, edgecolor="pink");
```

marker

```python
sns.stripplot(x="weight", y="origin", data = cars, linewidth=2, edgecolor="pink", alpha=0.5, size=8, jitter=0.25, marker="*");
```
