#Scatter Plots
1. relational plot 

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds= sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[diamonds.cut.isin(['Premium', 'Good'])&
                    diamonds.color.isin(['D','F','J'])
                    ].sample(n=100, random_state=22)
```

```python
sns.set_style("dark")
```

```python
sns.scatterplot(diamonds.carat, diamonds.price)
```

```python
sns.scatterplot(x="carat", y="price", data=  diamonds)
```

```python
sns.scatterplot(x="x", y="y", data=diamonds)
```

# 3 extra sematic to show more information

2. Style
3. Size

```python
sns.scatterplot(x="carat", y="price", hue="cut", data=  diamonds, palette=['purple','blue','pink','aqua','yellow'] )
```

continuous variable

```python
sns.scatterplot(x="carat", y="price",hue="depth", data=  diamonds)
```

```python
sns.scatterplot(x="carat", y="price",s=100, data=  diamonds)
```

```python
sns.scatterplot(x="carat", y="price",style="cut", data=  diamonds, s=100)
```

size

```python
sns.scatterplot(x="carat", y="price",size="cut", data=  diamonds , sizes=[150,200,10,20,15])
```

hue+style+size

```python
sns.scatterplot(x="carat", y="price",hue='cut',style="cut", data=  diamonds, s=100)
```

```python
sns.scatterplot(x="carat", y="price",hue='cut',style="color", data=  diamonds, s=100)
```

alpha

```python
sns.scatterplot(x="carat", y="price",alpha=0.6, data=  diamonds, s=100)
```

marker

```python
sns.scatterplot(x="carat", y="price",alpha=0.6, data=  diamonds, s=100, marker="*", edgecolor="pink")
```

# Regplot
1. Regression plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds.sample(n=200, random_state=44)
diamonds.shape
```

```python
sns.set_style("dark")
```

```python
sns.regplot(diamonds.carat, diamonds.price)
```

```python
sns.regplot(x="carat", y="price" , data=diamonds)
```

```python
sns.regplot(x="carat", y="price" , data=diamonds , fit_reg= False)
```

```python
sns.regplot(x="carat", y="price" , data=diamonds , scatter= False)
```

```python
sns.regplot(x="carat", y="price" , data=diamonds , ci= None)
```

Discrete variables

```python
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}

diamonds['cut_value'] = diamonds.cut.map(cut_map)
```

```python
diamonds.cut_value.value_counts()
```

Jitter

```python
sns.regplot(x="cut_value", y="price" , data=diamonds)
```

```python
sns.regplot(x="cut_value", y="price" , data=diamonds , x_jitter=0.1)
```

X_estimator


Model
1. Polynomial regressor

```python
sns.regplot(x="carat", y="price" , data=diamonds , order=2)
```

robust regression

```python
sns.regplot(x="carat", y="price" , data=diamonds , robust=True)
```

1. Order
2. Robust
3. Logistic
4. Lowess (locally weighted scatterplot smoothing)
5. confounding variable

Styling
1. marker 
2. scatter_kws = accepts a dictionary # style scatter of regplot
3. line_kws = accepts a dictionary # style line of regplot

```python
sns.regplot(x="carat", y="price" , data=diamonds , scatter_kws={'s':100, 'alpha':0.5, 'color': 'pink'})
```
